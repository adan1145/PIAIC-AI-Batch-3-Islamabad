###### TASK 1:
* Make a list <code>l1 = [1,2,3]</code>. Write it to the file element by element.
* Then update the file by appending list <code>l2 = [A,B,C]</code>
* After reading the modified file, the output should be:
     <code>[C,B,A,3,2,1]</code>

```python
l1 = [1,2,3]
with open('mm','w') as f:
    for e in l1:
        f.write(str(e)+'\n')
        
with open('mm','r+') as f:
    l3 = list(reversed(f.readlines()))
    l4 = list(reversed(['A','B','C']))
    f.seek(0)
    print(l3)
    print(l4)
    l5 = l4 + l3
    print(l5)
    for e in l5:
        f.write(str(e))
        f.write('\n')
```

**CSV Files**

```python
# Simplified version of spreadsheets/excel--> comma separated values
# Notice that the formatting in the Excel file has disappeared in the CSV file.
# A CSV file is nothing but text.
# Builtin CSV Module
# .reader method
import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    
print(cof)
```

```python
# Reading and storing every line in the form of list
# use list comprehension for line-wise
# each value as different value in list

import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    ppp = [l for l in cof]
    
print(ppp)
```

```python
# each value as different value in list
import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    ppp = [e for l in cof for e in l]
    
print(ppp)
```

```python
# without list comprehension, using +
import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    ppp = []
    for l in cof:
        ppp += l
    
print(ppp)
```

```python
# without list comprehension, using append
import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    ppp = []
    for l in cof:
        ppp.append(l)
    
print(ppp)
```

```python
# without list comprehension, using extend
import csv

with open("competitions.csv") as f:
    cof = csv.reader(f)
    ppp = []
    for l in cof:
        ppp.extend(l)
    
print(ppp)
```

```python
# Loading info. into them
# If no file, python creates it otherwise over-written
# you can't write data directly to the CSV file. You have to create a special object that handles the data
# delimeters can also be tabs, semicolons, pipes(|), carets(^)
import csv
with open("whatever.csv", "w", newline="") as f:
    d_h = csv.writer(f,delimeter = ",")
```

```python
# Loading info. using obj.writerow
import csv
with open("whatever.csv", "w", newline="") as f:
    data_handler = csv.writer(f,delimiter = ",")
    data_handler.writerow(["Year", "Event","Winner"])
    data_handler.writerow(["1995", "Best-Kept Lawn","None"])
    data_handler.writerow(["1999", "Gobstones","Welch National"])
```

```python
# appending data
with open("whatever.csv", "a", newline="") as f:
    data_handler = csv.writer(f,delimiter = ",")
    data_handler.writerow(["2006", "World Cup","Burkina Faso"])
    data_handler.writerow(["2011", "Butter Cup","France"])
    data_handler.writerow(["2012", "Coffee Cup","Brazil"])
```

```python
import json
```

```python
alphabet_letters = ["a", "b", "c"]
with open("alphabet_list.txt", "w") as f:
    f.write(alphabet_letters)
```

```python
# You can save a list in a CSV file, but a more straightforward approach is to use JSON.
with open("alphabet_list.json", "w") as f:
    json.dump(alphabet_letters, f)
```
